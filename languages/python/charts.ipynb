## Pie Chart

Shows a pie chart used within a [Working Genius](https://www.workinggenius.com/about) exercise.

Suggests that you have 2 working Geniuses, Competencies, and Frustrations out of the list of 6:

- Wonder
- Invention
- Discernment
- Galvanizing
- Enablement
- Tenacity

The plot below demonstrates a suggestion: when setting up your work schedule, block 15% of your time to dedicate on your frustrations, 25% of time on your competencies, and 60% on your geniuses.

Take a project you are working on. Analyze the tasks required to achieve a successful outcome, and allot your time for optimal execution!

```python
labels = 'Frustrations', 'Competencies', 'Geniuses'
sizes = [15, 25, 60]
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90);
```

```python
# add some sugar: https://youtu.be/0UIB9Y4OFPs?si=ktTwbfdzB9ETdY7X
# Red, Yellow, Green
colors = ['#FFB3BA', '#FFFFBA', '#BAFFC9']
# Explode Frustrations segment
explode = (0.1, 0, 0)  
# plot
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.0f%%',
    startangle=90,
    wedgeprops=dict(edgecolor='gray', linewidth=1)
)
# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')
plt.title('Suggested Time Spent at Work')
plt.tight_layout()
plt.show()
```

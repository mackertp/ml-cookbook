```java
public class Factorial {  
    static int factorial(int n){      
          if (n == 1)      
            return 1;      
          else      
            return(n * factorial(n-1));      
    }      
  
    public void test() {  
        System.out.println("Factorial of 5 is: " + factorial(5));  
    }  
}
```

```java
Factorial test = new Factorial();
test.test();
```

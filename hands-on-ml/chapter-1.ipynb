## Chapter 1: The Machine Learning Landscape
Specialized applications such as *Optical Character Recognition* (OCR) have leveraged *Machine Learning* for decades. It is not some science-fiction, terminator-esque field that it is sometimes passed off as. Machine learning is math that is used in a way to ensure that people are working smarter and not harder to obtain results. Take the example of a simple spam filter for your email inbox. While it is not *Westworld* level humanoid bots that have the potential to gain sentience, I'm sure you appreciate what that little filter does for you. There are many more examples that we could point to, but lets put machine learning in basic terms: it is the application of *methods that leverage data to improve performance on some set of tasks.*

Applications of machine learning are intertwined with the field of Artificial Intelligence. We have seen this field evolve to become ubiquitous with our everyday lives. You can ask *Alexa* or your *Google Home* to remind you of your calendar, *Spotify* will reccomend new playlists so that you stop looping on your Journey tracks from 1980, and *IBM* is on a mission to save the seals with an [autonomous sailboat](https://www.ibm.com/cloud/automation/mayflower-autonomous-ship).
